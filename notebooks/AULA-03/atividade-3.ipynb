## 3. Descrever os dados usando estatísticas de tendência central e dispersão

## Medidas de tendencia central e dispersão - DataFrame penguins

```python
df1_new = df1.describe()
#MODA 
moda = df1.mode()
df_moda = pd.DataFrame(moda)
df_moda
tabela_1 = df1_new.merge(df_moda, how='outer')
tabela_1
#INTERVALO INTERQUARTIL (q3 - q1)
q1 = df1.quantile(0.25)
q3 = df1.quantile(0.75)
intervalo = q3 - q1
df_intervalo = pd.DataFrame(intervalo)
transpose_intervalo = df_intervalo.transpose()
df_linha_intervalo = pd.DataFrame(transpose_intervalo)
tabela_2 = tabela_1.merge(df_linha_intervalo, how='outer')
tabela_2
#INTERVALO (max - min)
max = df1.max(numeric_only=True)
min = df1.min(numeric_only=True)
intervalo_2 = max - min
df_intervalo_2 = pd.DataFrame(intervalo_2)
transpose_intervalo_2 = df_intervalo_2.transpose()
df_linha_intervalo_2 = pd.DataFrame(transpose_intervalo_2)
tabela_3 = tabela_2.merge(df_linha_intervalo_2, how='outer')
tabela_3
#RENOMEANDO OS INDEX
tabela_4 = tabela_3.rename(index={0: 'Quantidades', 1: 'Média', 2:'Desvio Padrão', 3:'Mínimo', 4:'Q1', 5:'Mediana(Q2)', 6:'Q3', 7:'Máximo', 8:'Moda', 9:'Intervalo Interquartil', 10: 'Intervalo'})
tabela_4



```
